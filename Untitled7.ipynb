```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Load dataset
data3= pd.read_csv("forestfires.csv")
data3.head()
```

```python
data3.info()
```

```python
data3.describe()
```

```python
#convet categorical data to numbers
data3['month']=data3['month'].astype("category").cat.codes
data3["day"]=data3["day"].astype("category").cat.codes
data3.head
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data3=data3[["temp",'RH','wind','rain','area']])
plt.title("boxplot to detect outliers")
plt.show()
```

```python
q1=data3["area"].quantile(0.25)
q3=data3["area"].quantile(0.75)
IQR= q3-q1
data3=data3[(data3["area"]>=q1 - 1.5*IQR) & (data3["area"]<= q3 + 1.5*IQR)]
```

```python
sns.pairplot(data3[['temp','RH','wind','rain','area']])
plt.suptitle("pairploy of kewy variales")
plt.show()
```

```python
# feature distribution

features = ['temp','RH','wind','rain','area']
plt.figure(figsize=(12,7))
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sns.histplot(data3[col],kde=True)
  plt.title("distribution of col")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(data3.select_dtypes(include='number').corr(), annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Split Dataset (Features & Target)
x = data3[['temp','RH', 'wind','rain']]
y = data3[['area']]  # target variable

x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size = 0.2, random_state=42)
```

```python
# train linear regression model
linear_model=LinearRegression()
linear_model.fit(x_train, y_train)
y_pred_linear_model = linear_model.predict(x_test)
```

```python
# evaluation of linear regression model
mse_linear_model = mean_squared_error(y_test, y_pred_linear_model)
r2_linear_model = r2_score(y_test, y_pred_linear_model)
print("linera regression results::")
print("mean squared error:", mse_linear_model)
print("r2 score:",r2_linear_model)
```

```python
# train random forest model
rand_forest_model=RandomForestRegressor(n_estimators=100, random_state=43)
rand_forest_model.fit(x_train, y_train)
y_pred_linear_model=rand_forest_model.predict(x_test)
```

```python
# evaluate random forest model
mse_randforest = mean_squared_error(y_test, y_pred_linear_model)
r2_randforest = r2_score(y_test, y_pred_linear_model)

print("random forest result/;")
print("mean sqaured error:",mse_randforest)
print("r2 score:",r2_randforest)
```

```python
#visualixzation
models=["linear regression","random forest"]
mse=[mse_linear_model,mse_randforest]
r2=[r2_linear_model,r2_randforest]

plt.figure(figsize=(5,4))
sns.barplot(x=models, y=r2, palette = "viridis")
plt.title("model comparison (R2 score):")
plt.ylabel("R2 score")
plt.show()
```

```python
print(" Model Training Completed!")
if r2_randforest > r2_linear_model:
    print("Random Forest performed better with higher accuracy.")
else:
    print("Linear Regression performed better in this dataset.")
```
